```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer = 'Data_Frame_Of_Docking_Scores_And_SMILESs.csv')
df
```

```python
import numpy as np
import seaborn as sns

series_of_docking_scores = df['docking score']
the_mean = np.mean(series_of_docking_scores)
standard_deviation = np.std(series_of_docking_scores)

the_Axes = sns.distplot(
    a = series_of_docking_scores,
    bins = 100,
    kde_kws = {"label": "KDE"},
    hist_kws = {"label": "Histogram"}
)
the_Axes.set(
    xlabel = "Docking Score",
    ylabel = "Probability Density",
    title = (
        "Probability Density Distribution Of Docking Scores"
        + f"\nNumber Of Observations = {df.shape[0]}, Mean = {the_mean}, Standard Deviation = {standard_deviation}"
    )
)
import matplotlib.pyplot as plt
plt.legend()
plt.savefig("Probability_Density_Distribution_Of_Docking_Scores.png")
plt.show()
```
